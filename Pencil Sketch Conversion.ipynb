```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact, Button, VBox, HBox
from google.colab import files
from PIL import Image
import os
```

```python
image_path = "/content/IMG-20241218-WA0019.jpg"
image = cv2.imread(image_path)
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
gray_inv = 255 - gray
```

```python
def pencil_sketch(ksize,effect_type,blend_ratio):
    if ksize % 2 == 0:
        ksize += 1

    if effect_type == "Pencil Sketch (Grayscale)":
      blur = cv2.GaussianBlur(gray_inv, (ksize, ksize), 0)
      result = cv2.divide(gray, 255 - blur, scale=256)
      plt.figure(figsize=(12, 6))

      plt.subplot(1, 2, 1)
      plt.title("Original Image")
      plt.imshow(image)
      plt.axis('off')

      plt.subplot(1, 2, 2)
      plt.title(f"Pencil Sketch (ksize={ksize})")
      plt.imshow(result, cmap='gray')
      plt.axis('off')

      plt.show()

    elif effect_type == "Color Sketch":
        stylized = cv2.stylization(image, sigma_s=60, sigma_r=0.6)
        blended = cv2.addWeighted(image, blend_ratio, stylized, 1 - blend_ratio, 0)
        latest_sketch = blended

        plt.figure(figsize=(12, 6))
        plt.subplot(1, 2, 1)
        plt.title("Original Image")
        plt.imshow(image)
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.title(f"Realistic Color Sketch (blend={blend_ratio:.2f})")
        plt.imshow(blended)
        plt.axis('off')
        plt.show()
```

```python
interact(pencil_sketch,
         ksize=widgets.IntSlider(min=1, max=50, step=3, value=16),
         effect_type=widgets.Dropdown(
             options=["Pencil Sketch (Grayscale)", "Color Sketch"],
             value="Pencil Sketch (Grayscale)",
             description="Effect"
         ),
         blend_ratio=widgets.FloatSlider(min=0.0, max=1.0, step=0.1, value=0.5, description="Realism Blend")
        );
```
